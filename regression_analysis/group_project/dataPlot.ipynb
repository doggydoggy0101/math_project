```python
def plot(x, n):
    fig = plt.figure(figsize=(2*n,8))
    fig.add_subplot(211)
    sns.countplot(data=data, x=x, width=0.5, color="#2E86C1")
    fig.add_subplot(212)
    sns.barplot(data=data, x=x, y="score", width=0.5, color="#2E86C1")
    plt.show()
```
